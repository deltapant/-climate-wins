```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
path = r"C:\Users\suhas\Documents\careerfoundry\ml\datasets\weather-processed.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
# exclude DATE and MONTH columns
exclude_columns = ['DATE', 'MONTH']
columns_to_scale = [col for col in df.columns if col not in exclude_columns]

# Apply StandardScaler only to selected columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[columns_to_scale])

# Create a DataFrame from the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=columns_to_scale)

final_df = pd.concat([df[exclude_columns].reset_index(drop=True), scaled_df], axis=1)

# Step 6: Export to CSV
final_df.to_csv(r'C:\Users\suhas\Documents\careerfoundry\ml\datasets\weather-scaled.csv', index=False)
```
